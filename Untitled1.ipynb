```python
import requests
import pandas as pd
import time
```

```python
url = "https://pokeapi.co/api/v2/pokemon?limit=2000"
res = requests.get(url)
data = res.json()

pokemon_list = data["results"]
print("Total URLs:", len(pokemon_list))
```

```python
pokemon_data = []

for i, p in enumerate(pokemon_list):
    res = requests.get(p["url"])

    if res.status_code != 200:
        continue

    data = res.json()

    stats = {stat["stat"]["name"]: stat["base_stat"] for stat in data["stats"]}

    pokemon_data.append({
        "height": data["height"],
        "weight": data["weight"],
        "hp": stats.get("hp", 0),
        "attack": stats.get("attack", 0),
        "defense": stats.get("defense", 0),
        "sp_atk": stats.get("special-attack", 0),
        "sp_def": stats.get("special-defense", 0),
        "speed": stats.get("speed", 0),
        "type": data["types"][0]["type"]["name"]
    })

    # Avoid rate limit
    time.sleep(0.05)

    # Progress print
    if i % 100 == 0:
        print(f"Fetched {i} Pokémon")

df = pd.DataFrame(pokemon_data)
print("Final dataset size:", len(df))
df.head()
```

```python
df["total_stats"] = (
    df["hp"] + df["attack"] + df["defense"] +
    df["sp_atk"] + df["sp_def"] + df["speed"]
)

df["atk_ratio"] = df["attack"] / (df["sp_atk"] + 1)
df["def_ratio"] = df["defense"] / (df["sp_def"] + 1)
df["speed_ratio"] = df["speed"] / (df["total_stats"] + 1)
```

```python
X = df.drop(columns=["type"])
y = df["type"]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=300,
    max_depth=12,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
def predict_type(height, weight, hp, attack, defense, sp_atk, sp_def, speed):
    total = hp + attack + defense + sp_atk + sp_def + speed

    sample = pd.DataFrame([{
        "height": height,
        "weight": weight,
        "hp": hp,
        "attack": attack,
        "defense": defense,
        "sp_atk": sp_atk,
        "sp_def": sp_def,
        "speed": speed,
        "total_stats": total,
        "atk_ratio": attack/(sp_atk+1),
        "def_ratio": defense/(sp_def+1),
        "speed_ratio": speed/(total+1)
    }])

    pred = model.predict(sample)
    return le.inverse_transform(pred)[0]
```

```python
print(predict_type(4, 60, 35, 55, 40, 50, 50, 90))
```
